## Fit resolutions

```python
temp= AddPdfNorm(gaussian, gaussian, prefix=['g1','g2'])
print describe(temp)
dgaussian= rename(temp, ['x', 'mu1', 'sigma1', 'mu2', 'sigma2', 'f1'])
print describe(dgaussian)
```

```python
def double_gaussian(x, mu1, sigma1, mu2, sigma2, f1):
       return f1*exp(-0.5*(x-mu1)**2/sigma1**2)/(sqrt(2*pi)*sigma1)+(1-f1)*exp(-0.5*(x-mu2)**2/sigma2**2)/(sqrt(2*pi)*sigma2)
```

```python
def param_text(values, errors=None):
    txt= u''
    for k in values.keys():
        txt += u'%s = %5.4g'%(k, values[k])
        if errors!= None:  txt += u'±%5.4g'%errors[k]
        txt += u'\n'
    return txt
```

```python
def draw_fit(data, function, pars, errs, hbins, range):
    h,b,p= plt.hist(data, hbins, range, histtype='step', color='blue')
    A= (h*(b[1:]-b[:-1])).sum()
    bx= np.linspace(b[0], b[-1], len(b)*10)
    fx, fy= draw_pdf_with_midpoints(function, pars, bx, color='red', linewidth=2, scale=A)
    ptxt= param_text(pars, errs)
    ax.xlim(range)
    ax= plt.gca()
    #xlim, ylim= ax.get_xlim(), ax.get_ylim()
    #ax.text(xlim[0]+0.05*(xlim[1]-xlim[0]), ylim[0]+0.85*(ylim[1]-ylim[0]), ptxt)
    ax.text(0.05, 0.95, ptxt, horizontalalignment='left', verticalalignment='top', transform = ax.transAxes)
```

### Positron momentum resolution

```python
L_eP0 = UnbinnedLH(dgaussian, 1000*D.ep3[(D.ep3>0.050)&(D.ep3<0.055)&selection0])
L_eP1 = UnbinnedLH(dgaussian, 1000*D.ep3[(D.ep3>0.050)&(D.ep3<0.055)&selection1])
describe(L_eP0)
```

```python
vpars= dict(mu1=52, sigma1=0.2, mu2=52, sigma2= 1, f1=0.5)
epars= dict( error_mu1=0.1, error_sigma1=0.02, error_mu2=0.1, error_sigma2=0.02, error_f1=0.1) 
ranges= dict( limit_f1=(0,1), limit_sigma1=(0.1,0.4), limit_sigma2=(0.4,2.0))
pars= dict(vpars.items()+epars.items()+ranges.items())
print pars
```

```python
m_eP0 = Minuit(L_eP0, print_level=1, **pars)
m_eP0.set_up(0.5) #remember up is 0.5 for likelihood and 1 for chi^2
m_eP0.migrad();
```

```python
m_eP1 = Minuit(L_eP1, print_level=1, **pars)
m_eP1.set_up(0.5) #remember up is 0.5 for likelihood and 1 for chi^2
m_eP1.migrad();
```

```python
fig= plt.figure(figsize=(7,5))
ax= plt.subplot(111)
L_eP0.draw(m_eP0, bins=100, ax=ax)
plt.xlabel(r'$p_e$ (MeV)', fontsize='x-large')

fig= plt.figure(figsize=(7,5))
ax= plt.subplot(111)
L_eP1.draw(m_eP1, bins=100, ax=ax)
plt.xlabel(r'$p_e$ (MeV)', fontsize='x-large')
```
